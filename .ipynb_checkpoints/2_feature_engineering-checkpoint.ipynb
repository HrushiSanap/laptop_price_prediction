```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#df = pd.read_csv("data/clean_laptops_train.csv")
df = pd.read_csv("data/clean_laptops_test.csv")

df.head()
```

```python
df.manufacturer.unique()
```

```python
#one-hot encoding using pandas get_dummies
df = pd.get_dummies(df, columns=['manufacturer'], prefix='manufacturer', dtype = int)
```

```python
df.head()
```

```python
df.category.unique()
```

```python
#one-hot encoding using pandas get_dummies
df = pd.get_dummies(df, columns=['category'], prefix='category', dtype = int)
```

```python
df.head()
```

```python
df.os.unique()
```

```python
os_mapping = {
    'macOS': 'mac_os',
    'Mac OS': 'mac_os',
    'No OS': 'no_os',
    'Chrome OS': 'chrome_os'
}

# Replace the values
df['os'] = df['os'].replace(os_mapping)
```

```python
df.os.unique()
```

```python
#one-hot encoding using pandas get_dummies
df = pd.get_dummies(df, columns=['os'], prefix='os', dtype=int)
```

```python
df.head()
```

```python
df.cpu_provider.unique()
```

```python
#one-hot encoding using pandas get_dummies
df = pd.get_dummies(df, columns=['cpu_provider'], prefix='cpu_provider', dtype = int)
```

```python
df.head()
```

```python
df.gpu_provider.unique()
```

```python
#one-hot encoding using pandas get_dummies
df = pd.get_dummies(df, columns=['gpu_provider'], prefix='gpu_provider', dtype = int)
```
