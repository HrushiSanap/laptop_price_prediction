```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_train = pd.read_csv("data/clean_laptops_train.csv")
df_test = pd.read_csv("data/clean_laptops_test.csv")
```

```python
# eda with training data
df_train.describe()
```

```python
df_train.info()
```

```python
#box plot of numeric columns
#to check outliers
df_train_num = df_train.select_dtypes(include=['float64'])
fig, axes = plt.subplots(3, 3, figsize = (18, 12))
axes_flat = axes.flatten()

for i, col in enumerate(df_train_num):
    sns.boxplot(x = df_train_num[col], ax = axes_flat[i])

plt.show()
```

```python
sns.set_context('notebook')
sns.pairplot(df_train_num)
```

```python
#box plot of numeric columns test
#to check outliers
df_test_num = df_test.select_dtypes(include=['float64'])
fig, axes = plt.subplots(3, 3, figsize = (18, 12))
axes_flat = axes.flatten()

for i, col in enumerate(df_test_num):
    sns.boxplot(x = df_test_num[col], ax = axes_flat[i])

plt.show()
```

```python
sns.set_context('notebook')
sns.pairplot(df_test_num)
```
