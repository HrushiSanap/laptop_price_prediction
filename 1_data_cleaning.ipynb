```python
def parse_storage(storage_string):
    ssd_total = 0
    hdd_total = 0
    hybrid_total = 0

    parts = [part.strip() for part in storage_string.split("+")]

    for part in parts:
        match = re.search(r'(\d+(?:\.\d+)?)\s*(GB|TB|MB)', part, re.IGNORECASE)
        if not match:
            continue

        value = float(match.group(1))
        unit = match.group(2).upper()

        if unit == 'TB':
            value *= 1024
        elif unit == 'MB':
            value /= 1024

        part_lower = part.lower()
        if 'ssd' in part_lower:
            ssd_total += value
        elif 'hdd' in part_lower:
            hdd_total += value
        elif any(term in part_lower for term in ['flash', 'hybrid']):
            hybrid_total += value
            
    return {
        'SSD': int(ssd_total),
        'HDD': int(hdd_total),
        'Hybrid': int(hybrid_total)
    }
```

```python
storage_parsed = df['Storage'].apply(parse_storage)

# Create new columns
df['SSD'] = [item['SSD'] for item in storage_parsed]
df['HDD'] = [item['HDD'] for item in storage_parsed]
df['Hybrid'] = [item['Hybrid'] for item in storage_parsed]
```

```python
def parse_cpu(cpu_str):
    # Extract CPU provider (first word)
    cpu_provider = cpu_str.split()[0]
    speed_match = re.search(r'(\d+(?:\.\d+)?)\s*GHz', cpu_str, re.IGNORECASE)
    
    if speed_match:
        clock_speed = float(speed_match.group(1))
    else:
        clock_speed = None
    
    return {
        'CPU_Provider': cpu_provider,
        'Clock_Speed_GHz': clock_speed
    }
```

```python
cpu_parsed = df['CPU'].apply(parse_cpu)

# Create new columns
df['CPU_Provider'] = [item['CPU_Provider'] for item in cpu_parsed]
df['Clock_Speed_GHz'] = [item['Clock_Speed_GHz'] for item in cpu_parsed]
```
